```
#Data Wrangling
import pandas as pd 
import numpy as np
```

```
url1 = 'https://github.com/KathyZhang-upenn/CIS550-Project-NBA/raw/master/teams.csv'
teams = pd.read_csv(url1)
teams = teams.drop(columns=['ARENACAPACITY'])
# teams.isna().any().any()
# group_teams = teams.groupby(["TEAM_ID"])
# counts7 = group_teams.size()
# print(counts7)
# (counts7 == 1).all()
teams
```

```
url2 = 'https://github.com/KathyZhang-upenn/CIS550-Project-NBA/raw/master/players.csv'
players = pd.read_csv(url2)
players.isna().any().any()
# group_players = players.groupby(["TEAM_ID", "PLAYER_ID", "SEASON"])
# counts6 = group_players.size()
# print(counts6)
# (counts6 == 1).all()
players
```

```
url3 = 'https://github.com/KathyZhang-upenn/CIS550-Project-NBA/raw/master/games.csv'
games = pd.read_csv(url3)
games = games.drop(columns=['GAME_STATUS_TEXT'])
games.isna().any().any()
# group_games = games.groupby(["GAME_ID"])
# counts5 = group_games.size()
# print(counts5)
# (counts5 == 1).all()
games
```

```
url4 = 'https://github.com/KathyZhang-upenn/CIS550-Project-NBA/raw/master/games_details.csv'
game_details = pd.read_csv(url4)
game_details = game_details.drop(columns=['COMMENT', 'START_POSITION'])
game_details = game_details.fillna(0)
game_details.isna().any().any()
# group_game_details = game_details.groupby(["GAME_ID", "PLAYER_ID"])
# counts4 = group_game_details.size()
# print(counts4)
# (counts4 == 1).all()
game_details
```

```
url5 = 'https://github.com/KathyZhang-upenn/CIS550-Project-NBA/raw/master/ranking.csv'
ranking = pd.read_csv(url5)
ranking = ranking.drop(columns=['RETURNTOPLAY'])
ranking.isna().any().any()
# grouped_ranking = ranking.groupby(["TEAM_ID", "STANDINGSDATE"])
# counts3 = grouped_ranking.size()
# print(counts3)
# (counts3 == 1).all()
ranking
```

```
url6 = 'https://github.com/KathyZhang-upenn/CIS550-Project-NBA/raw/master/Seasons_Stats.csv'
season_stats = pd.read_csv(url6)
season_stats = season_stats.rename(columns={"Player": "PLAYER_NAME", "Tm": "TEAM", "Year": "YEAR", "Pos": "POSITION","Age": "AGE"})
season_stats = season_stats.drop(columns=['Unnamed: 0', 'blank2', 'blanl', '3PAr', 'ORB%', 'DRB%', 'TRB%','AST%','STL%','BLK%', 'TOV%','USG%', 'TS%', 'FTr', 'WS', 'WS/48', 'OBPM',	'DBPM',	'BPM', 'VORP'])
season_stats = season_stats[season_stats['YEAR'] > 1981.0] 
season_stats = season_stats.replace('\*','',regex=True)
season_stats = season_stats.fillna(0)
season_stats.isna().any().any()
season_stats['YEAR'] = season_stats["YEAR"].astype(int)
# grouped_season_stats = season_stats.groupby(["YEAR", "PLAYER_NAME", "TEAM", "POSITION"])
# counts = grouped_season_stats.size()
# print(counts)
# (counts == 1).all()
season_stats
```

```
url7 = 'https://github.com/KathyZhang-upenn/CIS550-Project-NBA/raw/master/NBA_Full_Salaries_2000-2019.csv'
salaries = pd.read_csv(url7)
salaries = salaries.rename(columns={"Name": "PLAYER_NAME", "Year": "YEAR", "Salaries": "SALARY", "Rank": "SEASON_RANK"})
salaries = salaries.drop(columns=['Unnamed: 0'])
salaries = salaries[salaries['YEAR'] > 2000] 
salaries = salaries[salaries['SALARY'].notna()]
salaries.isna().any().any()
# grouped_salaries = salaries.groupby(["YEAR", "PLAYER_NAME"])
# counts2 = grouped_salaries.size()
# print(counts2)
# (counts2 == 1).all()
salaries
```

```
teams.to_csv("teams.csv", index=False)
games.to_csv("games.csv", index=False)
players.to_csv("players.csv", index=False)
game_details.to_csv("game_details.csv", index=False)
ranking.to_csv("ranking.csv", index=False)
season_stats.to_csv("season_stats.csv", index=False)
salaries.to_csv("salaries.csv", index=False)

# from google.colab import files
# files.download("teams.csv")
# files.download("games.csv")
# files.download("players.csv")
# files.download("game_details.csv")
# files.download("ranking.csv")
# files.download("season_stats.csv")
# files.download("salaries.csv")
```
